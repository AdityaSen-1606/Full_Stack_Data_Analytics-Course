Que 2) In a quant test of the CAT Exam, the population standard deviation is known to be 100. A sample of 25 tests taken has a mean of 520. Construct an 80% CI about the mean.

Ans. Range of confidence interval can be derived by using formula CI = mean Â± (Z * (std_dev / sqrt(n))), where CI is confidence interval, Z is Z-score & for 80% confidence Z = 1.28155.

```python
import math

def confidence_interval(mean, std_dev, n, confidence_level):
    z_score = 1.28155
    margin_of_error = z_score * (std_dev / math.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

mean = 520
std_dev = 100
n = 25
confidence_level = 0.8

lower, upper = confidence_interval(mean, std_dev, n, confidence_level)

print("Confidence Interval (80%):", (lower, upper))
```

Que 3) A car believes that the percentage of citizens in city ABC that owns a vehicle is 60% or less. A sales manager disagrees with this. He conducted a hypothesis testing surveying 250 residents & found that 170 residents responded yes to owning a vehicle.
<br>a) State the null & alternate hypothesis.
<br>b) At a 10% significance level, is there enough evidence to support the idea that vehicle owner in ABC city is 60% or less.


**Ans.** Null Hypothesis: Residents who own vehicle is less than or equal to 60%
<br>Alternate Hypothesis: Residents who own vehicle is greater than 60%<br><br>
<br>B) Given Information: Sample size (n) = 250, Residents who own vehicle = 170
<br> according to one sample proportion test, p-hat = yes/sample size = 170/250 = 0.68 <br> using z-test at 10% significance level <br> z = (p-hat - p) / sqrt((p * (1 - p)) / n) => z = (0.68 - 0.60) / sqrt((0.60 * (1 - 0.60)) / 250)
<br> The Z-score value of 10% significance level is ~1.28, and calculating z of test and comparing with 10% significance value, z-score is greater than 1.28, hence we reject null hypothesis.  <br> So, according to test residents who own the vehicle is greater than 60%.

Que 4) What is the value of the 99 percentile? <br>
2,2,3,4,5,5,5,6,7,8,8,8,8,8,9,9,10,11,11,12

```python
data = [2,2,3,4,5,5,5,6,7,8,8,8,8,8,9,9,10,11,11,12]
total=len(data)
percentile=99

Rank = (percentile/100)*(total+1)
print("Rank: ",Rank)
Value_low_rank=data[int(Rank)-2]
Value_high_rank=data[int(Rank)-1]

Value = Value_low_rank+(Rank-int(Rank))*(Value_high_rank-Value_low_rank);
print("Value",Value)
```
